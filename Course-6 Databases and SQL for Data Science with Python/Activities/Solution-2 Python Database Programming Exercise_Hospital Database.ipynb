#### Exercise 1: Connect to your database server and print its version


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('/content/drive/MyDrive/Colab Notebooks/NPC/Course-6 /Hospitalinfo (1).db')

# Print SQLite database version
with conn:
    cur = conn.cursor()
    cur.execute('SELECT sqlite_version()')
    db_version = cur.fetchone()
    print('Database Version : ', db_version)

# # Close the connection
# conn.close()
```

#### Exercise 2: Fetch Hospital and Doctor Information using hospital Id and doctor Id


```python
import pandas as pd
# Fetch hospital and doctor information
with conn:
    cur = conn.cursor()
    cur.execute('''SELECT * FROM Hospital, Doctor WHERE Hospital.ID = Doctor.HospitalID;''')
    hospital_doctor_info = pd.DataFrame(cur.fetchall())
    hospital_doctor_info.columns = [description[0] for description in cur.description]
    display(hospital_doctor_info)
```

#### Exercise 3: Get the list Of doctors as per the given specialty and salary

```python
with conn:
    cur = conn.cursor()
    cur.execute(''' SELECT DocName, salary, Speciality FROM Doctor GROUP BY Speciality, salary ORDER BY salary;''')
    List_doctors = cur.fetchall()
    print(List_doctors)
```

#### Exercise 4: Get a list of doctors from a given hospital


```python
with conn:
    cur = conn.cursor()
    cur.execute('''SELECT DocName, Name FROM Doctor, Hospital WHERE Hospital.ID = Doctor.HospitalID AND Name = 'JP Hoplkins';''')
    hospital_doctor_info = cur.fetchall()
    print(hospital_doctor_info)
    
```
