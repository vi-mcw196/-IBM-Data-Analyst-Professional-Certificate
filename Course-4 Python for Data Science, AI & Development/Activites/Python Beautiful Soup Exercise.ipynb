####Q1. 
Write a Python program to find the title tags from a given html document.


```python
# import the BeautifulSoup and requests libraries
from bs4 import BeautifulSoup 
import requests
```

```python
# get the HTML content from a website
url = input('Enter the website URL: ')
response = requests.get(url)
html_content = response.content

# parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# find all the title tags in the HTML document
titles = soup.find_all('title')

# print the title tags
print('Title tags found in the HTML document:')
for title in titles:
    print(title.string)
```

####Q4.
Write a Python program to extract all the URLs from the webpage python.org that are nested within <li> tags from



```python
url = 'https://www.python.org/'

# make a request to the webpage
response = requests.get(url)

# parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# find all the <li> tags in the HTML content
li_tags = soup.find_all('li')

# extract the URLs that are nested within the <li> tags
urls = []
for li_tag in li_tags:
    a_tag = li_tag.find("a")
    if a_tag and a_tag.has_attr("href"):
        urls.append(a_tag["href"])

# print the extracted URLs
for url in urls:
    print(url)
```

####Q6.
Write a Python program to retrieve the HTML code of the title, its text, and the HTML code of its parent.

```python
# make a request to the webpage
url = input('Enter the website URL: ')
response = requests.get(url)

# parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# print the results
print("Title HTML:\n", soup.title)
print("\nTitle Text:\n", soup.title.text)
print("\nParent HTML:\n", soup.title.parent)
```

####Q7.
Write a Python program to create a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each HTML/XML tag and string.


```python
str1 = "<p>Some<b>bad<i>HTML Code</i></b></p>"
print("Original string:\n", str1)

soup = BeautifulSoup(str1, "xml")
print("\nFormatted Unicode string:")
print(soup.prettify())
```

####Q9
Write a Python program to wrap an element in the specified tag and create the new wrapper.

```python
soup = BeautifulSoup("<p>Python exercises.</p>", "html.parser")
print("Original Markup:")
print(soup.p.string.wrap(soup.new_tag("i")))
print("\nNew Markup:")
print(soup.p.wrap(soup.new_tag("div")))
```

####Q10
Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.


```python
html_content = '<a href="https://w3resource.com/">Python exercises<i>w3resource</i></a>'

soup = BeautifulSoup(html_content, "html.parser")
print("Original Markup:")
a_tag = soup.a
print(a_tag)
new_tag = soup.a.decompose()
print("After decomposing:")
print(new_tag)
```
