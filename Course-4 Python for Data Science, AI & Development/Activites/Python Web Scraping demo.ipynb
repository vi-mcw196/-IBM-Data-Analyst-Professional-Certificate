```python
#Step-1 : Importing libraries 

from bs4 import BeautifulSoup 
import requests
```

```python
#Step-2: Adding URL

link = 'http://ravindrababuravula.com/courses.php'
data = requests.get(link)
```

```python
#Step-3 : Creating a soup
#data.content

soup = BeautifulSoup(data.text, 'html.parser')
soup.prettify
```

```python
#Step-4 : Retrieve the Data

soup.title
```

```python
soup.find('h2', class_='center-title')
```

```python
soup.find('h2', class_='center-title').string
```

```python

soup.find('a', class_='title').Text
for i in soup.find_all('a', class_='title'):
  print(i.text)
```

```python
for d in soup.find_all('div', class_='description'):
  print(d.text)
```

```python
for d in soup.find_all('img', class_='img-responsive'):
  print(d)
```

```python
for d in soup.find_all('img', class_='img-responsive'):
  print(d.src)
```
