#### Q8 Write a python program to find the longest word in mytxt file

```python
with open('mytxt.txt', 'r') as f:
    text = f.read()

words = text.split()
longest_word = max(words, key=len)

print(f'The longest word in the file is: {longest_word}')
```

####Q10 Write a Python program to build a sign_Up and login forms for users.
####Your program should validate/authenticate entered credentials for existing users 
####or create new accounts for new users
####1 Create "Credentials.txt" file
####2 Add the following information into your Credentials.txt file <Name, Password>
    # Sarah,12ssr
    # Lisa,pass
    # Donna,D00N
####3 Create login function that takes two inputs (name,password)
    #3.1 Ask the user to enter their username and password
    #3.2 search if the names exist in the Credentials.txt file
        #3.2.1 if the name was found, check 
            #if the password is correct
        #3.2.2 return "Welcome <persons name> 
            #if the name and password were correct"
        #3.2.3 else return "wrong name or password"
    #3.2 if the name doesn't exist return "your name doesn't exist,
        #please create a new account"
####4 Create sign_Up function that takes two inputs(newName,newPassword)
    #Ask user to enter their name and password
    #append the new information into the Credentials.txt file, 
        #follow the same format "name,password"
####5 Test your program by logging in using the new account


```python
def login():
    # Ask the user for their username and password
    name = input("Enter your username: ")
    password = input("Enter your password: ")

    # Open the credentials file in read mode
    with open('Credentials.txt', 'r') as f:
        # Loop through each line in the file
        for line in f:
            # Split the line into the name and password
            stored_name, stored_password = line.strip().split(',')
            # Check if the entered name matches the stored name
            if name == stored_name:
                # Check if the entered password matches the stored password
                if password == stored_password:
                    return f"Welcome {name}!"
                else:
                    return "Wrong password."
        # If we get here, the name was not found in the file
        return "Your name doesn't exist, please create a new account."

def sign_up():
    # Ask the user for their new username and password
    name = input("Enter a new username: ")
    password = input("Enter a new password: ")

    # Open the credentials file in append mode
    with open('Credentials.txt', 'a') as f:
        # Write the new name and password to the file
        f.write(f"{name},{password}\n")

    print("User registered successfully!")

# Call the login function to start the program
if __name__ == "__main__":
    action = input("Do you want to sign up or login? ")
    if action == "login":
        result = login()
        print(result)
    elif action == "sign up":
        sign_up()
    else:
        print("Invalid input. Please enter 'sign up' or 'login'.")
```

```python
import hashlib

# Define a dictionary to store user data
user_data = {}

# Define a function to hash passwords
def hash_password(password):
    # Encode the password as UTF-8
    encoded_password = password.encode('utf-8')
    # Hash the password using SHA-256
    hashed_password = hashlib.sha256(encoded_password).hexdigest()
    # Return the hashed password
    return hashed_password

# Define a function to register new users
def sign_up():
    # Ask the user for their username and password
    username = input("Enter a username: ")
    password = input("Enter a password: ")
    # Hash the password
    hashed_password = hash_password(password)
    # Store the username and hashed password in the user_data dictionary
    user_data[username] = hashed_password
    print("User registered successfully!")

# Define a function to authenticate existing users
def login():
    # Ask the user for their username and password
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    # Check if the username exists in the user_data dictionary
    if username in user_data:
        # Hash the entered password
        hashed_password = hash_password(password)
        # Compare the hashed password with the stored hashed password
        if hashed_password == user_data[username]:
            print("Login successful!")
        else:
            print("Incorrect password.")
    else:
        print("Username does nnot exist.")

# Define a main function to handle user input
def main():
    # Loop until the user chooses to exit
    while True:
        # Ask the user if they want to sign up or login
        action = input("Do you want to sign up or login? ")
        # Call the appropriate function based on the user's choice
        if action == "sign up":
            sign_up()
        elif action == "login":
            login()
        else:
            print("Invalid input. Please enter 'sign up' or 'login'.")
        # Ask the user if they want to continue
        choice = input("Do you want to continue? (y/n) ")
        # Exit the program if the user chooses to quit
        if choice == "n":
            break

# Call the main function to start the program
if __name__ == "__main__":
    main()
```
